```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/mabuhannood/Downloads/Project/Orders_Cleaned.csv')
```

```python
# General Information
print("Dataset Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Basic Info:")
print(df.info())
print("Basic Statistics:")
print(df.describe())
```

### Distribution

```python
# Distribution of Numerical Features
numerical_cols = ['Order Amount', 'Rolls Ordered', 'Est. Extended Cost (Line)']
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True, bins=30, color="skyblue")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Distribution of Categorical Features
categorical_cols = ['Sales Region']
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette="coolwarm")
    plt.title(f"Countplot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()

# Correlation Heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

### Time-based Analysis


```python
# Plot Orders by Year
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Date Created Year', order=sorted(df['Date Created Year'].unique()), palette="viridis")
plt.title("Orders by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()

# Plot Orders by Month
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Date Created Month', order=range(1, 13), palette="viridis")
plt.title("Orders by Month")
plt.xlabel("Month")
plt.ylabel("Count")
plt.show()
```

### Relationship Analysis: 

```python
# Order Amount vs Rolls Ordered
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Rolls Ordered', y='Order Amount', color="teal")
plt.title("Order Amount vs Rolls Ordered")
plt.xlabel("Rolls Ordered")
plt.ylabel("Order Amount")
plt.show()

# Order Amount vs Est. Extended Cost (Line)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Est. Extended Cost (Line)', y='Order Amount', color="orange")
plt.title("Order Amount vs Est. Extended Cost (Line)")
plt.xlabel("Est. Extended Cost (Line)")
plt.ylabel("Order Amount")
plt.show()

print("Exploratory Data Analysis completed.")
```
